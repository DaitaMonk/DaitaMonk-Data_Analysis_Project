```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
pd.set_option("display.max_rows", None)
```

```python
# Reading the data
df = pd.read_csv('movies.csv')
```

```python
# A look at the data

df.head(200)
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# dropped rows with missing values
df = df.dropna()
```

```python
# After droping Na,s

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for columns

print(df.dtypes)
```

```python
# Budget and Gross Columns to integer

df[["budget", "gross"]] = df[["budget", "gross"]].astype("int64")
```

```python
# Are there any Outliers?

df.boxplot(column=['gross'])
```

```python
# check if there are Duplicated rows

df[df.duplicated()]
```

```python
# Order our Data a little bit to see

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
sns.regplot(x="gross", y="budget", data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})

plt.title("Budget vs Gross")

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
sns.regplot(x="score", y="gross", data= df, scatter_kws={"color":"orange"}, line_kws={"color":"Purple"})

plt.title("Score vs Gross")

plt.xlabel('Score')

plt.ylabel('Budget for Film')

plt.show()
```

```python
# Correlation Matrix between all numeric columns

df.corr(method ='pearson')
```

```python
df.corr(method ='kendall')
```

```python
df.corr(method ='spearman')
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

Using the for loop and ccat.codes method to assign a random numeric value for each unique categorical value

```python
df.head(200)
```

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head(200)
```

```python
df_numerized.corr(method='pearson')
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
