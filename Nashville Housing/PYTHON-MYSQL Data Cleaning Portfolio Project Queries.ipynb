Populate Property Address data

```python
pd.read_sql("""SELECT a.ParcelID, a.PropertyAddress, b.ParcelID, 
               b.PropertyAddress, IFNULL(a.PropertyAddress, b.PropertyAddress) 
               FROM NashvilleHousing a
               JOIN NashvilleHousing b 
               ON a.ParcelID = b.ParcelID
               AND a.UniqueID <> b.UniqueID
               WHERE a.PropertyAddress IS NULL""", con = engine)
```

```python
try:
    mycursor.execute("""UPDATE  NashvilleHousing a
                        JOIN NashvilleHousing b 
                        ON a.ParcelID = b.ParcelID
                        AND a.UniqueID <> b.UniqueID
                        SET a.PropertyAddress = IFNULL(a.PropertyAddress,b.PropertyAddress)
                        WHERE a.PropertyAddress IS NULL""")
    mydb.commit()
except:
    mydb.rollback()
```

Breaking out Address into Individual Columns (Address, City, State)

```python
pd.read_sql("""PropertyAddress
            FROM NashvilleHousing
            """, con = engine)
```

```python
pd.read_sql("""SELECT SUBSTRING(PropertyAddress, 1, POSITION(','IN PropertyAddress) -1 ) 
               AS Address, SUBSTRING(PropertyAddress, POSITION(','IN PropertyAddress) + 1 , 
               LENGTH(PropertyAddress)) AS Address
               FROM NashvilleHousing
               """, con = engine)
```

```python
try:
    mycursor.execute("""ALTER TABLE NashvilleHousing
                        ADD PropertySplitAddress VARCHAR(255); 
                        """)
    mycursor.execute("""Update NashvilleHousing
                        SET PropertySplitAddress = SUBSTRING(PropertyAddress, 1, 
                        POSITION(',' IN PropertyAddress) -1 ) """)
    
    mydb.commit()
except:
    mydb.rollback()
```

```python
try:
    mycursor.execute("""ALTER TABLE NashvilleHousing
                        ADD PropertySplitCity VARCHAR(255); 
                        """)
    mycursor.execute("""Update NashvilleHousing
                        SET PropertySplitCity = SUBSTRING(PropertyAddress,
                        POSITION(','IN PropertyAddress) + 1, LENGTH(PropertyAddress))""")
    
    mydb.commit()
except:
    mydb.rollback()
```

```python
pd.read_sql("""SELECT SUBSTRING_INDEX(OwnerAddress, ',', 1)
               ,SUBSTRING_INDEX(SUBSTRING_INDEX(OwnerAddress, ',', 2), ',', -1)
               ,SUBSTRING_INDEX(OwnerAddress, ',', -1)
               FROM NashvilleHousing""",
               con = engine)
```

```python
try:
    mycursor.execute("""ALTER TABLE NashvilleHousing
                        ADD OwnerSplitAddress VARCHAR(255); 
                        """)
    mycursor.execute("""Update NashvilleHousing
                        SET OwnerSplitAddress = SUBSTRING_INDEX(OwnerAddress, ',', 1) 
                        """)
    
    mycursor.execute("""ALTER TABLE NashvilleHousing
                        ADD OwnerSplitCity VARCHAR(255); 
                        """)
    mycursor.execute("""Update NashvilleHousing
                        SET OwnerSplitCity = SUBSTRING_INDEX
                        (SUBSTRING_INDEX(OwnerAddress, ',', 2), ',', -1) 
                        """)
    
    mycursor.execute("""ALTER TABLE NashvilleHousing
                        ADD OwnerSplitState VARCHAR(255); 
                        """)
    mycursor.execute("""Update NashvilleHousing
                        SET OwnerSplitState = SUBSTRING_INDEX(OwnerAddress, ',', -1) 
                        """)
    
    mydb.commit()
except:
    mydb.rollback()
```

Change Y and N to Yes and No in "Sold as Vacant" field

```python
pd.read_sql("""SELECT DISTINCT(SoldAsVacant), COUNT(SoldAsVacant)
                FROM NashvilleHousing
                GROUP BY SoldAsVacant
                ORDER BY 2""",
                con = engine)
```

```python
pd.read_sql("""SELECT SoldAsVacant
             , CASE WHEN SoldAsVacant = 'Y' THEN 'Yes'
               WHEN SoldAsVacant = 'N' THEN 'No'
               ELSE SoldAsVacant
               END
               FROM NashvilleHousing""",
                con = engine)
```

```python
try:
    mycursor.execute("""UPDATE NashvilleHousing
                        SET SoldAsVacant = CASE WHEN SoldAsVacant = 'Y' THEN 'Yes'
                        WHEN SoldAsVacant = 'N' THEN 'No'
                        ELSE SoldAsVacant
                        END""")
    
    mydb.commit()
except:
    mydb.rollback()
```
